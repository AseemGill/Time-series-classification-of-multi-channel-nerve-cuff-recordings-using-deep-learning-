```python
def plot_training_curve(path, save_path, save = False):
    """ Plots the training curve for a model run, given the csv files
    containing the train/validation error/loss.

    path - The base path of the csv files produced during training
    save_path - Where to save curves
    save - whether to save curves
    """
    
    
    import matplotlib.pyplot as plt
    
    fig, axs = plt.subplots(2)
    train_err = np.loadtxt("{}_train_acc.csv".format(path))
    val_err = np.loadtxt("{}_val_acc.csv".format(path))
    train_loss = np.loadtxt("{}_train_loss.csv".format(path))
    val_loss = np.loadtxt("{}_val_loss.csv".format(path))
    axs[0].set_title("Train vs Validation Accuracy")
    n = len(train_err) # number of epochs
    axs[0].plot(range(1,n+1), train_err, label="Train")
    axs[0].plot(range(1,n+1), val_err, label="Validation")
#     axs[0].xlabel("Epoch")
#     axs[0].ylabel("Error")
    axs[0].legend(loc='best')
#     plt.show()
    axs[1].set_title("Train vs Validation Loss")
    axs[1].plot(range(1,n+1), train_loss, label="Train")
    axs[1].plot(range(1,n+1), val_loss, label="Validation")
#     axs[1].xlabel("Epoch")
#     axs[1].ylabel("Loss")
    axs[1].legend(loc='best')
#     plt.show()
    
    if save == True:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
```

```python
# Model Checkpoints path and experiment to plot
base = "M:\Peripheral Nerve Studies\MCC Projects\Aseem G\Models\Pytorch\model_checkpoints\\"
kernel_set = "Test within Experiment"

for i in [2,3,4,5,6,7,8,9,10]:
    rat = "Rat " + str(i)
    base_dir = base + rat + "\\Experiment Name\\"
    
    for j in [1,2,3]:
        
        file_path = base_dir + rat + "Test within Experiment" + str(j)
        save_path = base_dir + "Fold_" + str(j) + "_plots"
        
#         print(file_path)
#         print(save_path)
        plot_training_curve(file_path, save_path, save = True)
        
        
```
